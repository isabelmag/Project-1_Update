# Step 4: Clean your training set. 

```python
print(conflict_train.isnull().sum())
```

There are a lot of nulls in this dataset! For each variable with NAs, what is the most appropriate way to change them? I invite bias in no matter what I choose - it's really about bias minimization and building the most realistic picture based on the data I do have.

Population total - I'm going to delete the observations that have nulls, because dropping 326 observations in a dataset of 12,000+ should not drastically alter the models, unless there is a meaningful reason why a country with a certain population wouldn't be included -- could dig deeper and determine if I need to weight certain countries with x population more
Rural and urban population and population density - I'm going to fill nulls with mean values


GINI index and presence of peacekeepers- have more nulls than not! I don't think this will be a useful feature, so I'm going to drop it
Arms imports and exports - going to drop for the moment but in the future I might want to create another feature, "Net Arms" that calculates the difference in imports and exports

```python
conflict_train = conflict_train.dropna(subset=["Population__total"])
```

```python
mean_rur = conflict_train["Rural_population"].mean() 
conflict_train["Rural_population"] = conflict_train["Rural_population"].fillna(mean_rur)
```

```python
mean_urb = conflict_train["Urban_population"].mean() 
conflict_train["Urban_population"] = conflict_train["Urban_population"].fillna(mean_urb)
```

```python
mean_density = conflict_train["Population_density__people_per_sq__km_of_land_area_"].mean() 
conflict_train["Population_density__people_per_sq__km_of_land_area_"] = conflict_train["Population_density__people_per_sq__km_of_land_area_"].fillna(mean_density)
```

```python
med_natincome = conflict_train["Adjusted_net_national_income_per_capita__current_US__"].median() 
conflict_train["Adjusted_net_national_income_per_capita__current_US__"] = conflict_train["Adjusted_net_national_income_per_capita__current_US__"].fillna(med_natincome)
```

```python
med_forces = conflict_train["Armed_forces_personnel__total"].median() 
conflict_train["Armed_forces_personnel__total"] = conflict_train["Armed_forces_personnel__total"].fillna(med_forces)
```

```python
med_death = conflict_train["Death_rate__crude__per_1_000_people_"].median() 
conflict_train["Death_rate__crude__per_1_000_people_"] = conflict_train["Death_rate__crude__per_1_000_people_"].fillna(med_death)
```

```python
med_gdp = conflict_train["GDP__current_US__"].median() 
conflict_train["GDP__current_US__"] = conflict_train["GDP__current_US__"].fillna(med_gdp)
```

```python
med_gdpcap = conflict_train["GDP_per_capita__current_US__"].median() 
conflict_train["GDP_per_capita__current_US__"] = conflict_train["GDP_per_capita__current_US__"].fillna(med_gdpcap)
```

```python
med_mil = conflict_train["Military_expenditure__current_USD_"].median() 
conflict_train["Military_expenditure__current_USD_"] = conflict_train["Military_expenditure__current_USD_"].fillna(med_mil)
```

```python
conflict_train = conflict_train.drop("Arms_exports__SIPRI_trend_indicator_values_", axis=1)
```

```python
conflict_train = conflict_train.drop("Arms_imports__SIPRI_trend_indicator_values_", axis=1)
```

```python
conflict_train = conflict_train.drop("GINI_index__World_Bank_estimate_", axis=1)
```

```python
conflict_train = conflict_train.drop("Presence_of_peace_keepers__number_of_troops__police__and_military_observers_in_mandate_", axis=1)
```
